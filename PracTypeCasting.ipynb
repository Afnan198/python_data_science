## Type Casting

### Type casting in python refers to the process of converting a value from one data type to another.This can be useful in various situation, when you need to perform operations betweeen different types or when you need to format data in a specific way. Python has several built in function for type casting

```python
int(6.5)
```

```python
b='1'
print(b)
```

```python
print(int(b))
```

```python
#print(b+1) its not possible because b is a string
```

```python
c=int(b)
print(c+1)
```

```python
d=26
print(str(d))
```
