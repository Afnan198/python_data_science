# Calculators

```python
a= 'first_num'
b='sec_num'

a=50
b=4
print("The value of {} + {} is".format(a, b),a+b)
print("The value of {} - {} is".format(a, b),a-b)
print("The value of {} * {} is".format(a, b),a*b)
print("The value of {} / {} is".format(a, b),a/b)
print("The value of {} % {} is".format(a, b),a%b)
print("The value of {} to the power {} is".format(a, b),a**b)
```
