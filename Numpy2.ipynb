```python
import numpy as np
list1=[1, 2, 3]
list2=[4,5,6]
list3=[7,8,6]
numpyarr=np.array([[list1,list2,list3]])
print(numpyarr)
```

## Numpy array attribute

```python
num_att=np.array([[1,2,3],[5,6,7]])
print(num_att)
```

```python
print("Shape:",num_att.shape)
print("Size:",num_att.size)
print("type:",type(num_att))
print("data type:",num_att.dtype)
print("Dimension:",num_att.ndim)
```

```python
arr1=np.array([1,2,3,4])
arr2=np.array([[[2,3,4],[2,4,5],[7,8,6]]])
print(arr1,"\n")
print(arr2,"\n")

print("For array 1:")
print("Shape:", arr1.shape)
print("Size:", arr1.size)
print("Dimension", arr1.ndim)
print("Data Type:", arr1.dtype)

print("\nFor array 2:")
print("Shape:", arr2.shape)
print("Size:", arr2.size)
print("Dimension", arr2.ndim)
print("Data Type:", arr2.dtype)
```

```python
#Zero array
zero=np.zeros((3,4))
print(zero)
# here 3 is the row no 4 is the column no.how many rows and column would be 
# needed
```

```python
#One array
one=np.ones((3,4))
print(one)
# here 3 is the row no 4 is the column no.how many rows and column would be 
# needed
```

```python
#full array
full1=np.full((3,4),11)
print(full1)
# here 3 is the row no 4 is the column no.how many rows and column would be 
# needed and 11 is what number we want to print
```

```python
#one array using full array
full2=np.full((3,4),1)
print(full2)
```

```python
#zero array using full array
full2=np.full((3,4),0)
print(full2)
```
